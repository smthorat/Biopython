## Iterating over the records in a sequence file

```python
from Bio import SeqIO
record_iterator = SeqIO.parse("/Volumes/Jagannath/Programming/ls_orchid.fasta.txt", "fasta")
print(record_iterator)
```

It is important to understand the difference between parse and read 

Use SeqIO.parse when you need to handle multiple records from a file, especially when working with large datasets.
Use SeqIO.read when you are certain your file contains exactly one record and you want to read it in a straightforward manner.

So now the question is, why do we need to use next function because it iterate through one by one and it does not read all the sequences at one times that saves the memory it is more useful for large files when you say first record it only fetch the first record then you said second record third record, and one by one it can fetch the records, which is more memory efficien

```python
first_record = next(record_iterator)
print(first_record.id)
```

```python
print(first_record.description)
```

```python
second_record = next(record_iterator)
print(second_record.id)
print(second_record.description)
```

```python
# Note if you try to use next() and there are no more records, you will special StopIteration exception
```
